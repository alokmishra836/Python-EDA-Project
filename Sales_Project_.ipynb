```python
df['Order Date'] = pd.to_datetime(df['Order Date'],errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'],errors='coerce')
df
```

```python
monthly_sales = df.groupby('Month_Year')['Sales'].sum()

monthly_sales.plot(kind='line', figsize=(12,5))
plt.title("Monthly Sales Trend")
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar',color='pink')
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='barh', figsize=(10,6),color='red')
plt.title("Sales by Sub-Category")
plt.show()
```

#  Business Insights â Superstore Sales Analysis

---

## 1. What Categories Perform Best?

When we look at the data overall, one thing becomes pretty clear:

- **Technology products are the strongest performers**, especially in terms of profit.
- **Office Supplies** stay consistently stable â not extremely high, but reliable.
- **Furniture is tricky** â it often shows good sales numbers, but the profit doesnât match up.

### What this tells us:
Itâs important to remember that *high sales donât always mean high profit*. Some categories simply cost more to serve or are heavily discounted.

### So what should the business do?
- Focus more on growing the **Technology category**
- Revisit pricing and cost structure in **Furniture**
- Keep Office Supplies as a steady, dependable category

---

## 2. What is actually causing losses?

This is where things get interesting.

- Heavy **discounts are eating into profits**
- Some Furniture items (like Tables and Bookcases) repeatedly show losses
- Even products with strong sales sometimes end up losing money

### What this tells us:
The business is probably using discounts too aggressively without checking the impact on profit.

### What should be done?
- Be more careful with discount strategies
- Review pricing for products that consistently lose money
- Consider removing or redesigning products that donât make financial sense

---

## 3. Which region is performing the best?

Not all regions are equal in performance:

- The **West region clearly stands out** â strong sales and good profit balance
- The **East region sells well**, but profit is not equally strong
- The **Central region is underperforming** in terms of profitability

### What this tells us:
Some regions are generating revenue, but not efficiently turning it into profit.

### Business actions:
- Invest more in the **West region**
- Improve efficiency in the **Central and East regions**
- Investigate why costs are higher in weaker regions

---

## 4. What do customer buying patterns look like?

Customer behavior is quite uneven:

- A small group of customers contributes a large share of total sales
- Many customers only buy once and donât return
- High-value customers are extremely important for revenue stability

### What this tells us:
This follows the classic **80/20 rule (Pareto principle)** â a small number of customers drive most of the business.

### What should be done?
- Build loyalty programs for top customers
- Try to convert one-time buyers into repeat customers
- Re-engage inactive customers with targeted offers

---

## 5. How do products perform overall?

Looking at product-level behavior:

- A few products generate most of the revenue
- Some products sell a lot but still donât make profit
- Many products contribute very little overall

### What this tells us:
The product portfolio is not evenly efficient â some items are carrying the business while others add little value.

### Business actions:
- Focus marketing on top-performing products
- Remove or improve loss-making products
- Avoid holding unnecessary inventory for low performers

---

## 6. What about discounts?

Discounts are a double-edged sword:

- They do increase sales volume
- But they often reduce profit significantly

### What this tells us:
Discounting is being used for growth, but not always in a profitable way.

### Recommendation:
- Use discounts more strategically, not universally
- Set limits depending on product category and margin

---

## 7. Overall takeaway

If we step back and look at everything together:

- Profit is driven more by **smart pricing and product selection** than just sales volume
- **Technology + West region** are the strongest growth areas
- Discounts need better control
- Customer base is concentrated â so retention is very important

---

## Final Thought

This analysis shows that the business is doing well in terms of sales, but there is still a lot of room to improve profitability.

Small changes in pricing, discount strategy, and customer focus can significantly improve overall performance.
